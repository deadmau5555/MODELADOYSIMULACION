**2.8.3**(Calibrating the Euler method) The goal of this problem is to test the
Euler method on the initial value problem x' =−x , x (0) = 1.

```
import numpy as np
import matplotlib.pyplot as plt
import math
def f(xn):
  return -1*xn

def euler(N,h,xo):
  x=np.zeros(N)
  x[0]=xo
  for n in range(1,N):
    x[n]=x[n-1]+f(x[n-1])*h
  return x
```

**2.8.4** Redo Exercise 2.8.3, using the improved Euler method.

**h=1**

```
# Valores de simulación
t0=0
tN=10
N= 10
t= np.linspace(t0,tN,N)
x_imp=np.zeros(N)
h= (tN-t0)/N  

# Condicion inicial
x_imp[0]= 1

for i in range(N-1):
  x_imp[i+1]=x_imp[i] + h * ( -1*(x_imp[i]+(h*-x_imp[i])/2) )
plt.figure(figsize=(16, 5), dpi=90)
plt.plot(t,x_imp,'-bo',label='euler mejorado')
plt.ylabel('x(t)')
plt.xlabel('t')
plt.legend()
plt.grid()
plt.show()
```

**2.8.5** Redo Exercise 2.8.3, using the Runge–Kutta method.

**h=1**

```
# Valores de simulación
t0=0
tN=10
N= 10
t= np.linspace(t0,tN,N)

h= (tN-t0)/N
x_4rk=np.zeros(N)
# Condicion inicial
x_4rk[0]= 1



def g(x_):
  return -1*x_

for n in  range (N-1):
  k1=g(x_4rk[n]   )   
  k2=g(x_4rk[n]+ k1/2)
  k3=g(x_4rk[n]+ k2/2)
  k4=g(x_4rk[n]+ k3  )

  x_4rk[n+1]=x_4rk[n]+ h/6*(k1+ 2*k2+ 2*k3 + k4)


plt.plot(t,x_4rk,'-bo',label='RK4')
plt.ylabel('x(t)')
plt.xlabel('t')
plt.legend()
plt.grid()
```
